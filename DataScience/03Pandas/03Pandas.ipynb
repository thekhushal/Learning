```python
# Concatenation in pandas ---> using this method we can combine two dataframes vertically or horizontally
```

```python
data1 = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 40], 
    "Address": ["New York", "Los Angeles", "Chicago", "Houston"], 
    "Qualification": ["BSc", "MSc", "PhD", "MBA"]}
df1 = pd.DataFrame(data1, index=[0, 1, 2, 3])
df1
```

```python
data2 = {
    "Name": ["Eve", "Frank", "Grace", "Heidi"],
    "Age": [28, 33, 38, 45],
    "Address": ["Phoenix", "Philadelphia", "San Antonio", "San Diego"],
    "Qualification": ["BA", "MA", "PhD", "MBA"]}
df2 = pd.DataFrame(data2, index=[4, 5, 6, 7])
df2
```

```python
a = [df1, df2]
pd.concat(a) # Vertical concatenation
```

```python
data1 = {
    "Name" : ["Alice", "Bob", "Charlie", "David"],
    "Age" : [25, 30, 35, 40], 
    "Address" : ["New York", "Los Angeles", "Chicago", "Houston"], 
    "Qualification" : ["BSc", "MSc", "PhD", "MBA"]
    "ID" : [1, 2, 3, 4]} 
df1 = pd.DataFrame(data1, index=[0, 1, 2, 3])
df1

data2 = {
    "Name" : ["Eve", "Frank", "Grace", "Heidi"],
    "Age" : [28, 33, 38, 45],
    "Address" : ["Phoenix", "Philadelphia", "San Antonio", "San Diego"],
    "Qualification" : ["BA", "MA", "PhD", "MBA"],
    "Salary" : [50000, 60000, 70000, 80000]}
df2 = pd.DataFrame(data2, index=[0, 1, 2, 3])
df2
```

```python
res2 = pd.concat([df1, df2], axis=1, join = "outer") # Horizontal concatenation
res2
```
