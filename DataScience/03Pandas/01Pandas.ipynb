```python
import pandas as pd
```

```python
a = pd.Series([1, 2, 3, 4, 5])  # Corrected 'series' to 'Series'
a = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
a = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'], name='numbers')
a = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'], name='numbers', dtype='int64')
print(a)
```

```python
b = {
    "emp_id": [1,2,3,4,5],
    "Name": ['a', 'b', 'c', 'd', 'e'], 
    "Department": ["IT", 'IT', "Sales", "Sales", "HR"],
    "working hours":[8,8,9,8,9],
    "salary": [20000, 25000, 20000, 25000, 30000]
}
print(b)
```

```python
df = pd.DataFrame(b)
print(df)
```

```python
# How to export data to csv
df.to_csv("emp_data.csv")
```
