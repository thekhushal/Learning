# Pandas drop() Complete Guide

```python
import pandas as pd

df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'marks': [85, 72, 91, 65],
    'city': ['Delhi', 'Mumbai', 'Pune', 'Jaipur']
})

df
```

## Drop Rows

```python
df.drop(1)
```

## Drop Columns

```python
df.drop(columns='city')
```
