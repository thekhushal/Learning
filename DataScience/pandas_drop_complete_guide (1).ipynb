# Complete Guide to `drop()` in pandas

This notebook covers:

- Dropping rows
- Dropping columns
- `axis`, `index`, `columns`
- `inplace=True`
- `errors='ignore'`
- Duplicate labels
- MultiIndex edge cases
- Conditional dropping
- Common mistakes
- `del` vs `drop` vs `pop`
- Performance considerations


```

import pandas as pd
import numpy as np

df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'marks': [85, 42, 91, 33],
    'city': ['Delhi', 'Mumbai', 'Pune', 'Jaipur']
})

df
```

## 1. Dropping Rows (Default axis=0)

```

df.drop(1)
```

## 2. `drop()` Uses Labels, NOT Positions

```

df2 = pd.DataFrame(
    {'name': ['A', 'B', 'C']},
    index=[10, 20, 30]
)

print(df2)

# Drops label 20
df2.drop(20)
```

## 3. Dropping Columns Using `axis=1`

```

df.drop('city', axis=1)
```

## 4. Preferred Modern Style: `columns=`

```

df.drop(columns='city')
```

## 5. Dropping Multiple Columns

```

df.drop(columns=['name', 'city'])
```

## 6. Dropping Multiple Rows

```

df.drop([1, 3])
```

## 7. Preferred Explicit Style: `index=` and `columns=`

```

df.drop(index=[0, 2])
```

## 8. Using `inplace=True`

```

temp = df.copy()

temp.drop(columns='city', inplace=True)

temp
```

## 9. IMPORTANT: `inplace=True` Returns None

```

temp = df.copy()

result = temp.drop(columns='city', inplace=True)

print(result)
```

## 10. `errors='ignore'`

```

df.drop(columns='missing_column', errors='ignore')
```

## 11. Duplicate Labels Edge Case

```

dup_df = pd.DataFrame(
    {'A': [1, 2, 3]},
    index=[0, 0, 1]
)

print(dup_df)

# Drops ALL rows labeled 0
dup_df.drop(0)
```

## 12. Conditional Dropping

```

df.drop(df[df['marks'] < 40].index)
```

## 13. MultiIndex Row Dropping

```

arrays = [
    ['A', 'A', 'B', 'B'],
    [1, 2, 1, 2]
]

index = pd.MultiIndex.from_arrays(arrays, names=('group', 'id'))

multi_df = pd.DataFrame({'value': [10, 20, 30, 40]}, index=index)

print(multi_df)

# Drop all rows where level 0 == 'A'
multi_df.drop(index='A', level=0)
```

## 14. MultiIndex Columns Edge Case

```

columns = pd.MultiIndex.from_tuples([
    ('math', 'midterm'),
    ('math', 'final'),
    ('science', 'final')
])

multi_col_df = pd.DataFrame(
    np.random.randint(0, 100, (3, 3)),
    columns=columns
)

print(multi_col_df)

# Drop top-level 'math' columns
multi_col_df.drop(columns='math', level=0)
```

## 15. `drop()` vs `dropna()`

```

nan_df = pd.DataFrame({
    'A': [1, np.nan, 3],
    'B': [4, 5, np.nan]
})

print(nan_df)

print("\nUsing dropna():")
print(nan_df.dropna())
```

## 16. Series.drop()

```

s = pd.Series([10, 20, 30], index=[0, 1, 2])

s.drop(1)
```

## 17. Cannot Drop by Value Directly

```

# This drops index label 'Alice', NOT rows containing 'Alice'

temp = df.copy()
temp.index = ['Alice', 'Bob', 'Charlie', 'David']

temp.drop('Alice')
```

## 18. Correct Way to Drop by Value

```

df[df['name'] != 'Alice']
```

## 19. Empty Result Edge Case

```

df.drop(df.index)
```

## 20. Duplicate Columns Edge Case

```

dup_cols = pd.DataFrame(
    [[1, 2, 3]],
    columns=['A', 'A', 'B']
)

print(dup_cols)

# Drops ALL 'A' columns
dup_cols.drop(columns='A')
```

## 21. `del` vs `drop`

```

temp = df.copy()

del temp['city']

temp
```

## 22. `pop()`

```

temp = df.copy()

removed = temp.pop('city')

print("Removed column:")
print(removed)

print("\nRemaining DataFrame:")
print(temp)
```

## 23. Performance Consideration

```

# drop() usually creates a new DataFrame unless inplace=True

new_df = df.drop(columns='city')

print("Original DataFrame ID:", id(df))
print("New DataFrame ID:", id(new_df))
```

## 24. Most Common Beginner Mistakes

```

# Mistake 1:
# Thinking drop(0) means first row

# Mistake 2:
# Chaining inplace operations

# WRONG:
# df.drop(columns='city', inplace=True).head()

# Because inplace=True returns None

# Mistake 3:
# Forgetting axis when dropping columns

# WRONG:
# df.drop('city')

# RIGHT:
df.drop(columns='city')
```
