```python
import numpy as np
import pandas as pd
```

```python
# Creating sample series

subs = pd.read_csv('../04CampusX_DataSet/datasets-session-16/subs.csv').squeeze()
# subs.head()

vk = pd.read_csv('../04CampusX_DataSet/datasets-session-16/kohli_ipl.csv',index_col='match_no').squeeze()
# vk.head()
```

# **9 Series Methods**

```python
# 1. Head | Tail | Sample
vk.head() # Top 5 rows
vk.head(10) # Top x rows; x being num in the parenthesis

vk.tail() # Bottom 5 rows
vk.tail(8) # Bottom x rows; x being num in the parenthesis

vk.sample() # 5 random rows
vk.sample(2) # x random rows; x being num  in the parenthesis
```

```python
# 2. value counts
vk.value_counts().head()
```

```python
# 3. sort value

vk.sort_values() # Default order of sorting is ascending on index
vk.sort_values(ascending=False)
vk.sort_values(na_position='first')

# To make permanent changes to series
# vk.sort_values(ascending=False, inplace=True) # # ERROR -> because Series is immutable, we cannot change the values of a Series. We can only create a new Series with the sorted values.
vk = vk.sort_values(ascending=False) 
```

```python
# 4. sort index
vk.sort_index(ascending=False)

# To make permanent changes to series
# vk.sort_index(ascending=False, inplace=True) 
```

```python
# 5 count
vk.count()
```

```python
# 6 sum | prod 
vk.prod()
vk.sum()
```

```python
# 7. mean | median | mode | std | var

print(subs.mean())
print(vk.median())
print(subs.mode()) # returns a series whereas rest rturn single number as mode can have multiple values
print(subs.std())
print(vk.var())
```
