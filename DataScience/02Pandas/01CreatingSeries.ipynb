# **3 Ways of Making Series**
1. List
2. Dictionary
3. CSV

```python
# 1. Using List

# Internally, pandas Series constructor is defined as follows:
# pd.Series(data=None, index=None, dtype=None, name=None, copy=None)

marks = [67,57,89,100,0]
subjects = ['maths','english','science','hindi', 'hindi']

# Creating series
pd.Series(data = marks, index = subjects, name = "Marks") # Creates a series with a name

# the same series can also be written as 
pd.Series(marks, subjects, np.int64, "Marks")
```

```python
# 2. Creating a Series from Dictionary

marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}

marks_series = pd.Series(data = marks,name='nitish ke marks')
marks_series
# NOTE: although pd.series allow Non-unique index values in index, but creating a series with non-unique index isnt posible with dictionary as keys of a dictionary are always unique
# NOTE: WORK-AROUND you you passs a index list with dictionary data, pandas will overwrite the index with the list provided for index as shown bellow
# pd.series(data=marks, index=["maths", "hindi", "science", "hindi"], name="just trying out something") 
```
