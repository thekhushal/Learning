# **3 Ways of Making Series**
1. List
2. Dictionary
3. CSV

```python
# 1. Using List

# Internally, pandas Series constructor is defined as follows:
# Series(data=None, index=None, dtype=None, name=None, copy=None)

marks = [67,57,89,100,0]
# pd.Series(marks)

subjects = ['maths','english','science','hindi', 'hindi']

# Series with index
# pd.Series(data=marks, index = subjects) # Creates a series with sub as index

# Series with name
pd.Series(marks, subjects, name = "Uske Marks") # Creates a series with a name

# When we are using the sequence of parameters, we need to follow the order of parameters as defined in the constructor. If we are using the keyword arguments, then we can use any order of parameters.

```

```python
# 2. Creating a Series from Dictionary

marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}

marks_series = pd.Series(marks,name='nitish ke marks')
marks_series
# NOTE: since dict cant have same index (hindi twice) but list can, when we crete series with list series can have index repeting, when when we create them with dict, they cant
```
