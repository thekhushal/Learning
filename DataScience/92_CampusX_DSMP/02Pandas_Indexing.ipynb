
# Pandas `.loc` and `.iloc` Explained

This notebook explains:

- What `.loc` is
- What `.iloc` is
- Differences between them
- Common real-world usage patterns
- Important edge cases


````python

import pandas as pd

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["Delhi", "Mumbai", "Pune"]
}, index=["a", "b", "c"])

df
````


## `.loc` — Label Based Selection

`.loc` works with:

- Row labels
- Column labels


````python

# Single row
df.loc["a"]
````

````python

# Single value
df.loc["b", "Age"]
````

````python

# Multiple rows
df.loc[["a", "c"]]
````

````python

# Multiple rows and columns
df.loc[["a", "b"], ["Name", "Age"]]
````

````python

# Slicing with .loc (end INCLUDED)
df.loc["a":"c"]
````


## `.iloc` — Integer Position Based Selection

`.iloc` works with integer positions.

Row positions:

- a → 0
- b → 1
- c → 2


````python

# Single row
df.iloc[0]
````

````python

# Single value
df.iloc[1, 1]
````

````python

# Multiple rows and columns
df.iloc[[0, 2], [0, 2]]
````

````python

# Slicing with .iloc (end EXCLUDED)
df.iloc[0:2] # slicing rows
````

````python
df.iloc[0:2, 0:2] # slicing rows and columns together
````

````python
df.iloc[:, 0:2] # slicing columns for all rows
````

````python
# when you need specific columns
df.iloc[0:2].loc[:, ["Name", "City"]]
````

````python
df.iloc[0:2].iloc[:, [0, 2]]
````


## Boolean Filtering with `.loc`

Very common in real-world data analysis.


````python

df.loc[df["Age"] > 25]
````

````python

df.loc[df["Age"] > 25, ["Name", "City"]]
````


## Setting Values


````python
df
````

````python

# Using .loc
df.loc["a", "Age"] = 26

# Using .iloc
df.iloc[1, 1] = 31

df
````


## Integer Index Confusion

One of the most important edge cases.


````python

df2 = pd.DataFrame(
    {"A": [10, 20, 30]},
    index=[100, 101, 102]
)

df2
````

````python

# Label-based
df2.loc[100]
````

````python

# Position-based
df2.iloc[0]
````


## Negative Indexing


````python

# Last row using iloc
df.iloc[-1]
````


`.loc[-1]` only works if `-1` exists as an actual index label.



## Common Industry Usage Patterns


````python

# Filtering rows
df.loc[df["Age"] > 25]

# Selecting columns
df.loc[:, ["Name", "Age"]]

# First column by position
df.iloc[:, 0]

# First 2 rows
df.iloc[:2]
````


# Quick Summary

| Feature | `.loc` | `.iloc` |
|---|---|---|
| Based on | labels | positions |
| Uses names | Yes | No |
| Slice end included | Yes | No |
| Negative indexing | label-based | positional |
| Common for filtering | Yes | Less common |

## Mental Model

```python
.loc[label]
.iloc[position]
```
