### Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

### Importing Pandas

```python
import numpy as np
import pandas as pd
```

### Series from lists

```python
# string
country = ['India','Pakistan','USA','Nepal','Srilanka']

pd.Series(country)
```

```python
# integers
runs = [13,24,56,78,100]

runs_ser = pd.Series(runs)
```

```python
# custom index
marks = [67,57,89,100]
subjects = ['maths','english','science','hindi']

pd.Series(marks,index=subjects)
```

```python
# setting a name
marks = pd.Series(marks,index=subjects,name='Nitish ke marks')
marks
```

```python
# Roll number
roll_no = [1,2,3,4,5]
names = ['Nitish','Rahul','Satyarth','Rohit','Satyam']
pd.Series(names, index=roll_no)
```

```python
# sequence of parameters
# Internally, pandas Series constructor is defined as follows:
# Series(data=None, index=None, dtype=None, name=None, copy=None)
# pd.Series(names, index=roll_no, name='Students') # WORKS
pd.Series(names, roll_no, object, 'Students') # WORKS
# pd.Series(names, roll_no, 'Students', object) # ERROR
```

```python
# pd.Series(data=names, name = 'Students', index=roll_no)
pd.Series(name = 'Students', index=roll_no, data=names, dtype=object)


# BASICALLY:
# When we are using the sequence of parameters, we need to follow the order of parameters as defined in the constructor. If we are using the keyword arguments, then we can use any order of parameters.

```
