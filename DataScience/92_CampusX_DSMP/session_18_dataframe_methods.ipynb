# **3. Rank ---> (Series)**

```python
batsman = pd.read_csv('../04CampusX_DataSet/datasets-session-17/batsman_runs_ipl.csv')
batsman.head()
```

```python
batsman['batsman_run'].rank(ascending=False)
# Creates a series assigning ranks to the batsman runs in descending order
# each index of the series corresponds to the index of the batsman in the dataframe
# ranks are calculated based on the batsman_run column of the dataframe
# since ascending=False, rank 1 ---> highest run and so on
# two batsman with same runs will get the same rank and the next rank will be skipped
# for example, if two batsman have the highest run of 100, they will both get rank 1 and the next rank will be 3
```

```python
# batsman['batting_rank'] => Creates a in DataFrame column named 'batting_rank' since no such column exists in the dataframe already.
# batsman['batsman_run'].rank(ascending=False) => Ranks the batsman runs in descending order and returns a series of ranks

batsman['batting_rank'] = batsman['batsman_run'].rank(ascending=False)
batsman.sort_values('batting_rank')
```

```python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
def euclidean(row):
  pt_A = row['1st point']
  pt_B = row['2nd point']

  return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5
```

```python
points_df['distance'] = points_df.apply(euclidean,axis=1)
points_df
```
