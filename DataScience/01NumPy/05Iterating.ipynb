```python
import numpy as np
```

```python
a = np.arange(50).reshape(10,5)
a
```

```python
print(a[[0, 3, 4, 5, 6, 9]])
# another way to do the same thing is using np.take() function
```

```python
for i in a:
    print(i)
```

```python
for i in a:
    for j in i:
        print(j, end=" ")
```

```python
for i in np.nditer(a):
    print(i, end=" ")
```
