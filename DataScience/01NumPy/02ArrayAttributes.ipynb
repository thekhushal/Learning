### **Array Attributes**

```python
import numpy as np
```

```python
# Creating 3 different arrays with different dimensions and dtypes
a1 = np.arange(10, dtype=np.int32) 
a2 = np.arange(12,dtype=np.float64).reshape(3,4)
a3 = np.arange(8, dtype=np.float64).reshape(2,2,2)

print(a1, end="\n\n")
print(a2, end="\n\n")
print(a3)
```

```python
# Dimension
print(f'a1 dimension: {a1.ndim}')
print(f'a2 dimension: {a2.ndim}')
print(f'a3 dimension: {a3.ndim}')
```

```python
# shape
print(f'a1 shape: {a1.shape}')
print(f'a2 shape: {a2.shape}')
print(f'a3 shape: {a3.shape}')
```

```python
# Size
print(f'a1 size: {a1.size}')
print(f'a2 size: {a2.size}')
print(f'a3 size: {a3.size}')
```

```python
# Itemsize
print(f'a1 itemsize: {a1.itemsize} bytes')
print(f'a2 itemsize: {a2.itemsize} bytes')
print(f'a3 itemsize: {a3.itemsize} bytes')
```

```python
# Dtype
print(f'a1 dtype: {a1.dtype}')
print(f'a2 dtype: {a2.dtype}')
print(f'a3 dtype: {a3.dtype}')
```

```python
# Changing dtype ---> astype() method
cd = a2.astype('complex128')

print(f'a2 dtype: {a2.dtype}')
print(f'cd dtype: {cd.dtype}')

print(f'cd: \n{cd}')
```
