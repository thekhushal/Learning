### **Indexing and Slicing**

```python
import numpy as np
```

```python
a = np.arange(12).reshape(3, 4)
a
```

```python
a[:2, 1::]
```

```python
a3 = np.arange(27).reshape(3,3,3)
print(a3)
```

```python
a3[1]
```

```python
a3[0, 1]
```

```python
a3[1, : , 1]
```

```python
a3[2, 1:, 1:]
```

```python
a3[::2, 0, ::2]
```
