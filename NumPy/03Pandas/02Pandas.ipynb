```python
# loc() and iloc() methods
# df.loc[row_range, column_name] for label-based indexing ----> last index value will be included
# df.iloc[row_range, column_index] for positional indexing ----> last index value will be excluded
df.loc[0]  # Access the first row using loc
df.iloc[0]  # Access the first row using iloc
```

```python
df.loc[0:2]  # Access the first two rows using loc (inclusive of last index)
df.iloc[0:2]  # Access the first two rows using iloc (exclusive of last index)

df.loc[2:5, ['age', 'gender', 'had_covid']]  # Access rows 2 to 5 and specific columns using loc
df.iloc[2:5, [0, 1, 3]]  # Access rows 2 to 5 and specific columns using iloc (using column indices
```

```python
# how can we check missing values in the DataFrame?
df.isnull().sum()  # Check for missing values in each column and sum them up
a = df 
a = a.dropna()  # Drop rows with any missing values
a.isnull().sum()  # Check for missing values again after dropping rows
a.shape
```

```python
# How can we fill the missing values in the DataFrame?
#  df['column_name'] ----> Numerical ----> column mean fill
# df['column_name'].fillna(value, inplace=True)  # Fill missing values in a specific colum

# df['col'] ----> Categorical ----> column mode fill
```

```python
df['fever'] = df['fever'].fillna(df['fever'].mean())  # Fill missing values in 'fever' column with mean
df.shape
df['has_covid'].mode()
df['has_covid'].value_counts()
```
