```python
# entirely vs code generated code
# This code performs sentiment analysis on text data using TextBlob and visualizes the results with Pandas and Seaborn.
def textblob_sentiment(text):
    """
    Calculate sentiment polarity and subjectivity using TextBlob.
    
    Parameters:
    text (str): The input text to analyze.
    
    Returns:
    tuple: A tuple containing polarity and subjectivity.
    """
    blob = TextBlob(text)
    return blob.sentiment.polarity, blob.sentiment.subjectivity

# Example usage
text = "I love programming in Python! It's so much fun."
polarity, subjectivity = textblob_sentiment(text)
print(f"Polarity: {polarity}, Subjectivity: {subjectivity}")

# Example DataFrame
data = {
    'text': [
        "I love programming in Python! It's so much fun.",
        "I hate bugs in my code. They are so annoying.",
        "Debugging is a challenging but rewarding process.",
        "I enjoy learning new libraries and frameworks.",
        "Sometimes I feel frustrated when things don't work as expected."
    ]
}

df = pd.DataFrame(data)
# Apply sentiment analysis to the DataFrame
df[['polarity', 'subjectivity']] = df['text'].apply(lambda x: pd.Series(textblob_sentiment(x)))

# Display the DataFrame with sentiment scores
print(df)   
# Plotting sentiment scores
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='polarity', y='subjectivity', hue='polarity', palette='coolwarm', s=100)
plt.title('Sentiment Analysis using TextBlob')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
plt.axhline(0.5, color='red', linestyle='--', label='Subjectivity Threshold')
plt.axvline(0, color='green', linestyle='--', label='Neutral Polarity')
plt.legend()
plt.grid()
plt.show()
# End of code
```
