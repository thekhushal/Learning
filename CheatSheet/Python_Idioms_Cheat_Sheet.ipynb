# Python One-Liners and Pythonic Idioms

A quick reference with examples.

## 1. Ternary Operator

```python
age = 20
status = "Adult" if age >= 18 else "Minor"
status
```

```python
ages = [13, 24, 43]

status = ["Adult" if age >= 18 else "Minor" for age in ages]
print(status)
```

```python
import numpy as np

ages = np.array([13, 24, 43])

status = np.where(ages >= 18, "Adult", "Minor")
print(status)
```

## 2. List Comprehension

```python
squares = [i*i for i in range(10)]
squares
```

## 3. List Comprehension with Condition

```python
evens = [i for i in range(10) if i % 2 == 0]
evens
```

## 4. Conditional Expression in List Comprehension

```python
result = ["Even" if i % 2 == 0 else "Odd" for i in range(10)]
result
```

## 5. Dictionary Comprehension

```python
d = {i:i**2 for i in range(5)}
d
```

## 6. Set Comprehension

```python
words = ["apple","banana","kiwi","pear"]
unique_lengths = {len(word) for word in words}
unique_lengths
```

## 7. Generator Expression

```python
sum(i*i for i in range(1000))
```

## 8. Multiple Assignment

```python
a,b = 10,20
print(a,b)
```

## 9. Swap Variables

```python
a,b = 10,20
a,b = b,a
print(a,b)
```

## 10. enumerate()

```python
names=["Alice","Bob","Charlie"]
for i,name in enumerate(names):
    print(i,name)
```

## 11. zip()

```python
names=["Alice","Bob","Charlie"]
ages=[25,30,35]
for name,age in zip(names,ages):
    print(name,age)
```

## 12. any() and all()

```python
numbers=[1,2,3,101]
print(any(x>100 for x in numbers))
print(all(x>0 for x in numbers))
```

## 13. lambda

```python
square = lambda x: x*x
print(square(5))

data=[{"name":"Alice","age":30},{"name":"Bob","age":20}]
print(sorted(data,key=lambda x:x['age']))
```

## 14. Unpacking

```python
x,y,z=[10,20,30]
print(x,y,z)
a,_,c=(1,2,3)
print(a,c)
a,*middle,b=[1,2,3,4,5]
print(a,middle,b)
```

## 15. Walrus Operator

```python
lines=iter(["hello","world","quit"])
while (line := next(lines)) != "quit":
    print(line)
```

## 16. Chained Comparisons

```python
x=5
print(0 < x < 10)
```

## 17. _ for Unused Variables

```python
for _ in range(3):
    print('Hello')
```

## 18. *args and **kwargs

```python
def add(*numbers):
    return sum(numbers)
print(add(1,2,3,4))
```

## 19. dict.get()

```python
d={'a':1}
print(d.get('a',0))
print(d.get('b',0))
```

## 20. np.where()

```python
import numpy as np
arr=np.array([2,5,7,9])
print(np.where(arr>5,'Yes','No'))
```
